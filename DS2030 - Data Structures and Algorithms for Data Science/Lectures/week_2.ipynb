<b> Big-Oh notation:</b>

- Def: given functions _f(n)_ and _g(n)_ , we say that _f(n)_ is _O(g(n))_ if there are positive constants c and n<sub>0</sub> such that <br> _f(n)_ $\le$ _cg(n)_ for _n_ $\ge$ _n<sub>0</sub>_
- Ex: 2n + 10 is O(n)... <br> 2n + 10 $\le$ cn <br> (c-2)n $\ge$ 10 <br> n $\ge$ 10/(c-2) <br> Pick c=2 and n<sub>0</sub>=10
- Ex: n<sup>2</sup> is <b>not</b> O(n)...
- Ex: 3n<sup>3</sup> + 2n<sup>2</sup> + 5 is O(n<sup>3</sup>)

Basically Big-Oh notation gives us an upper bound for the function _f(x)_ for values of x $\ge$ n<sub>0</sub> ...<br>
We can use this to rank functions according to their growrtth rates...<br>
If functions have same growth rates, then f(n) is O(g(n)) and g(n) is O(f(n))! <br>

```python
# -------- mutual recursion --------- #

def isOdd(n):
    if n == 1:
        return True
    elif n==0:
        return False
    else:
        return isEven(n-1)

def isEven(n):
    if n==2 or n==0:
        return True
    else:
        return isOdd(n-1)

print(isOdd(2))
print(isEven(4))
print(isEven(3))
print(isOdd(3))
```

**_Binary Search_**

```python
def bin_search(arr, target, min1, max1):
    if min1>max1:
        return (False, None)
    else:
        mid = (min1+max1)//2
        if target == arr[mid]:
            return (True, mid)
        elif target <arr[mid]:
            return bin_search(arr, target, min1, mid-1)
        else:
            return bin_search(arr, target, mid+1, max1)
```
